```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
data_path = '../data/raw'
```

```python
# Explore dataset structure
classes = os.listdir(data_path)
print("Classes found : ", classes)
```

```python
# Count images per class
class_counts = {}
for cls in classes:
  images = os.listdir(os.path.join(data_path,cls))
  class_counts[cls] = len(images)
```

```python
# Plot distribution
plt.bar(class_counts.keys(), class_counts.values())
plt.savefig('../results/class_distribution.png')
plt.show()
```

```python
# Display sample images

fig, axes = plt.subplots(6, 4, figsize=(15, 15))

for ax, cls in zip(axes.flat, classes):
    
    # Get first image from class folder
    img_path = os.path.join(data_path, cls, os.listdir(os.path.join(data_path, cls))[0])
    
    # Open image
    img = Image.open(img_path)
    
    # Show image
    ax.imshow(img)
    
    # Set title
    ax.set_title(cls)
    
    # Remove axis
    ax.axis("off")

plt.tight_layout()
plt.savefig('../results/sample_images.png')
plt.show()
```

```python
# Check image dimensions
dif_dimention = []
for cls in classes:
  print(f"Classes {cls} dimentions")
  for i in range(3):
    img_path = os.path.join(data_path, cls, os.listdir(os.path.join(data_path,cls))[i])
    img_open = Image.open(img_path)
    if img_open.size not in dif_dimention:
      dif_dimention.append(img_open.size)
    print(f"Image Size of {img_path.split('/')[-1]} :- {img_open.size}")
  print('\n')
print("Different Image Dimensions:", dif_dimention)
```
